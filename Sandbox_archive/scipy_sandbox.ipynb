```python
import numpy as np
import matplotlib.pyplot as pp
import scipy
from scipy.fft import *
import math
import cmath

%matplotlib inline
```

```python
x = np.linspace(0,10,2000)
signal = np.sin(x*2*math.pi) + np.sin(2*math.pi*3*x)
```

```python
pp.plot(x,signal)
```

```python
fft_signal = fft(signal)
```

```python
pp.plot(np.linspace(-1,1,2000),2/2000*abs(fft_signal))
```
